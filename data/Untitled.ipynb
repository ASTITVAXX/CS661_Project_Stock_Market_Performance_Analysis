```python
import pandas as pd
import pickle as pk
import numpy as np
```

```python
with open("stock_info.pkl",'rb') as f:
    data = pk.load(f)
data
```

```python
dat_without_none = [data[s] for s in data if data[s][0]!=None]
dat_without_none[:5]
```

```python
len(dat_without_none)
```

```python
sectors = list(set([d[2] for d in dat_without_none]))
sectors.remove('')
sectors
```

```python
dfzips = pd.read_csv("uszips.csv",dtype=str)
```

```python
states = dfzips['state_name'].unique()
states
```

```python
def zip_to_state(zip_code):
    ans = dfzips[dfzips['zip']==zip_code]['state_name']
    return ans.iloc[0]
```

```python
ans = dfzips[dfzips['zip']=='01752']['state_name']
ans.iloc[0]
```

```python
cols = ["State"]
cols.extend(sectors)
geodf = pd.DataFrame(columns=cols)
```

```python
geodf
```

```python
zeros = [0 for i in range(len(sectors))]
for s in states:
    row = [s]
    row.extend(zeros)
    geodf.loc[s]=row
```
